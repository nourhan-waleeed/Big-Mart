# Big Mart 
# The aim of this data science project is to build a predictive model and find out the sales of each product.

# Data Description

Item_Weight: The weight of the item.

Item_Fat_Content: The fat content category of the item (e.g., Low Fat, Regular).

Item_Visibility: The visibility of the item on the shelves.

Item_Type: The type or category of the item (e.g., Dairy, Soft Drinks, Meat, etc.).

Item_MRP: The Maximum Retail Price of the item.

Outlet_Identifier: Identifier for the outlet or store where the item is sold.

Outlet_Establishment_Year: The year when the outlet was established.

Outlet_Size: The size of the outlet (e.g., Small, Medium, High).

Outlet_Location_Type: The location type of the outlet (e.g., Tier 1, Tier 2, Tier 3).

Outlet_Type: The type of outlet (e.g., Supermarket Type1, Supermarket Type2, Grocery Store).

Item_Outlet_Sales: The sales value of the item at the respective outlet.

```python
# hot-deck imputation for training data

columns_to_impute = ['Outlet_Size', 'Item_Weight']

# Generate a list of indices with missing values 
missing_indices = train[train[columns_to_impute].isnull().any(axis=1)].index.tolist()

#iterate through missing_indices and perform hot-deck imputation
for index in missing_indices:
    #choose a random index that is not null 
    valid_indices = train[train[columns_to_impute].notnull().all(axis=1)].index.tolist()
    random_index = random.choice(valid_indices)
    
    # Impute with the values from the randomly selected index
    imputed_values = train.loc[random_index, columns_to_impute]
    train.loc[index, columns_to_impute] = imputed_values
```

```python
# hot-deck imputation for Testing data

missing_indices = test[test[columns_to_impute].isnull().any(axis=1)].index.tolist()

#iterate through missing_indices and perform hot-deck imputation
for index in missing_indices:
    #choose a random index that is not null 
    valid_indices = test[test[columns_to_impute].notnull().all(axis=1)].index.tolist()
    random_index = random.choice(valid_indices)
    
    # Impute with the values from the randomly selected index
    imputed_values = test.loc[random_index, columns_to_impute]
    test.loc[index, columns_to_impute] = imputed_values
```

```python
# Check if missing values have been imputed
missing_values_after_imputation = train[columns_to_impute].isnull().sum()
print("Missing Values in The Training Data after Hot-Deck Imputation:\n", missing_values_after_imputation)
```

# MI Score

```python
# Compute mutual information scores 
mi_selector = SelectKBest(score_func=mutual_info_regression, k='all')
mi_selector.fit(X_train, y_train)

mi_scores = mi_selector.scores_
feature_names= X_train.columns

# Sort the features by their MI scores 
sorted_features = sorted(zip(feature_names, mi_scores), key=lambda x: x[1], reverse=True)
print("\nMI Scores for All Cars:")
for feature, score in sorted_features:
    print(feature, ':', score)
```
